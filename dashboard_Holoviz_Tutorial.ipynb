```python
df = df.fillna(0)

# https://numpy.org/doc/stable/reference/generated/numpy.where.html
df['gdp_per_capita'] = np.where(df['population'] != 0, df['gdp']/df['population'],0)
```

```python
# Make DataFrame Pipeline Interactive - NOT SURE what this does
# presumably makes data frame so no longer STATIC

# holoviz docs explanation - https://holoviz.org/tutorial/Interactive_Pipelines.html

idf = df.interactive()
```

### CO2 emission over time by continent

```python
# define panel widgets
# https://panel.holoviz.org/user_guide/Widgets.html
year_slider = pn.widgets.IntSlider(name = 'Year slider', start =1750, end = 2020, step =5, value = 1850)
year_slider
```

```python
# Radio buttons for CO2 measures
# https://panel.holoviz.org/api/panel.widgets.html?highlight=radio#panel.widgets.RadioButtonGroup

yaxis_co2 = pn.widgets.RadioButtonGroup(
    name='Y axis', 
    options=['co2', 'co2_per_capita',],
    button_type='success'
)

yaxis_co2
```

```python
continents = ['World', 'Asia', 'Oceania', 'Europe', 'Africa', 'North America', 'South America', 'Antarctica']

# filtering interactive dataframe by 2 columns using BOOLEAN LOGIC
# where year <= slider value AND where the country in the shortlist of continents

# then ORDERS SUBSETS of the dataframe by country and then by year 
co2_pipeline = (
    idf[
        (idf.year <= year_slider) &
        (idf.country.isin(continents))
    ]
    .groupby(['country', 'year'])[yaxis_co2].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='year')  
    .reset_index(drop=True)
)

# groupby .... mean() - https://pandas.pydata.org/docs/reference/api/pandas.core.groupby.GroupBy.mean.html
# presumably accounts for double entries. If 2 entires for World 1750 CO2 levels, then take the mean!!

# sort by year explicitly sorts the SUBSETS by year!!
```

```python
#co2_pipeline

# plot dataframe to SCREEN
co2_plot = co2_pipeline.hvplot(x = 'year', by='country', y=yaxis_co2,line_width=2, title="CO2 emission by continent")
co2_plot
```

### Table - CO2 emission over time by continent

```python
# docs - search pagination and TABULATOR widget
# conjecture .pipe connected the the WIDGET 
# passed in through the brackets co2_pipeline object (which is a graph)

co2_table = co2_pipeline.pipe(pn.widgets.Tabulator, pagination='remote', page_size = 10, sizing_mode='stretch_width') 
co2_table
```

```python
#### CO2 vs GDP scatterplot

# country, year, gdp per capita == 1 unique COMBINATION of these 3 variables
# hence why sort by year after defining the UNIQUE COMBINATIONS
co2_vs_gdp_scatterplot_pipeline = (
    idf[
        (idf.year == year_slider) &
        (~ (idf.country.isin(continents)))
    ]
    .groupby(['country', 'year', 'gdp_per_capita'])['co2'].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='year')  
    .reset_index(drop=True)
)
```

```python
co2_vs_gdp_scatterplot = co2_vs_gdp_scatterplot_pipeline.hvplot(
    x='gdp_per_capita', 
    y='co2', 
    by='country', 
    size=80, kind="scatter", 
    alpha=0.7,
    legend=False, 
    height=500, 
    width=500
)
co2_vs_gdp_scatterplot
```

### Bar chart with CO2 sources by continent



```python
# Creating RADIO BUTTON with 4 options
# yaxis_co2_source is a STRING, it seems. String selected using the GUI!!
yaxis_co2_source = pn.widgets.RadioButtonGroup(
    name='Y axis', 
    options=['coal_co2', 'oil_co2', 'gas_co2'], 
    button_type='success'
)

# defining continents
continents_excl_world = ['Asia', 'Oceania', 'Europe', 'Africa', 'North America', 'South America', 'Antarctica']

# defining another interactive object using the idf. Again linking it to the
# slider and other radio button

co2_source_bar_pipeline = (
    idf[
        (idf.year == year_slider) &
        (idf.country.isin(continents_excl_world))
    ]
    .groupby(['year', 'country'])[yaxis_co2_source].sum()
    .to_frame()
    .reset_index()
    .sort_values(by='year')  
    .reset_index(drop=True)
)
```

```python
co2_source_bar_plot = co2_source_bar_pipeline.hvplot(
    kind='bar', 
    x='country', 
    y=yaxis_co2_source, 
    title='CO2 source by continent'
)
co2_source_bar_plot
```

### Create webdashboard

- Holoviz has pre-defined ("supported templates") templates
- Fastlist template used here

- See template docs - https://panel.holoviz.org/user_guide/Templates.html

```python
#Layout using Template
template = pn.template.FastListTemplate(
    title='World CO2 emission dashboard', 
    sidebar=[pn.pane.Markdown("# CO2 Emissions and Climate Change"), 
             pn.pane.Markdown("#### Carbon dioxide emissions are the primary driver of global climate change. It’s widely recognised that to avoid the worst impacts of climate change, the world needs to urgently reduce emissions. But, how this responsibility is shared between regions, countries, and individuals has been an endless point of contention in international discussions."), 
             pn.pane.Markdown("## Settings"),   
             year_slider],
    main=[pn.Row(pn.Column(yaxis_co2, 
                           co2_plot.panel(width=700), margin=(0,25)), 
                 co2_table.panel(width=500)), 
          pn.Row(pn.Column(co2_vs_gdp_scatterplot.panel(width=600), margin=(0,25)), 
                 pn.Column(yaxis_co2_source, co2_source_bar_plot.panel(width=600)))],
    accent_base_color="#88d8b0",
    header_background="#88d8b0",
)

# main section has 2 rows with charts in them!! 
# template.show()
template.servable();
```
